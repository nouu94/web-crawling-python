### 엑셀 파일 만들기
- 함수로 작성해보면서, 함수 작성법과 활용법 이해하기

````python
import openpyxl

def write_excel_template(filename, sheetname, listdata):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width = 100
    excel_sheet.column_dimensions['B'].width = 20
    
    if sheetname != '':
        excel_sheet.title = sheetname
    
    for item in listdata:
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
````

````python
import openpyxl

def write_excel_template(filename, sheetname, listdata) :
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width = 80
    excel_sheet.column_dimensions['B'].width = 20
    
    if sheetname != '' :
        excel_sheet.title = sheetname
        
    for item in listdata :
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
````

### 크롤링해서 엑셀 파일까지 만들기
- 리스트 안에 리스트(각 행을 나타냄)가 들어가야 함

````python
import requests
from bs4 import BeautifulSoup

fmkorea_afreeca_best_list = list()

for page_num in range(1, 16):
    res = requests.get('https://www.fmkorea.com/index.php?mid=afreecatv&sort_index=pop&order_type=desc&page=' + str(page_num))
    soup = BeautifulSoup(res.content, 'html.parser')

    data = soup.select('div.fm_best_widget._bd_pc > ul > li > div')
    for item in data:
        title = item.select_one('ul > li > div > h3 > a').get_text().strip()
        if title.count('[') >= 2 :
            title = title.replace('[', '(', 1).split('[')[0].replace('(', '[')
        else :
            title = title.split('[')[0]
            
        kinds = item.select_one('ul > li > div > div:nth-child(4) > span > a').get_text()
        times = item.select_one('ul > li > div > div:nth-child(5) > span.regdate').get_text().strip()
        writer = item.select_one('ul > li > div > div:nth-child(5) > span.author').get_text().strip(' / ')
        
        info = [title, kinds, times, writer]
        fmkorea_afreeca_best_list.append(info)
            
#         product_info = [product_name.get_text().strip(), product_date.get_text()]
#         product_lists.append(product_info)
````

````python
write_excel_template('펨코 앞갤 인기글 현황.xlsx', '20년12월5일', fmkorea_afreeca_best_list)
````
